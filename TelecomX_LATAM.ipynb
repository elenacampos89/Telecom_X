# Extracci贸n

````python
import pandas as pd

# URL RAW del archivo JSON en GitHub
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"

# Cargar los datos desde la URL
df = pd.read_json(url)

# Mostrar las primeras filas para verificar
df.head()
````

# Transformaci贸n

````python
# Mostrar informaci贸n general del DataFrame (columnas, no nulos, tipos)
print(" Informaci贸n general del DataFrame:")
df.info()

# Tipos de datos individuales
print("\n Tipos de datos por columna:")
print(df.dtypes)

# Ver las primeras filas para entender la estructura
print("\n Primeras filas del dataset:")
display(df.head())

# Ver valores 煤nicos de la variable objetivo 'Churn'
print("\n Distribuci贸n de la variable 'Churn':")
print(df['Churn'].value_counts(dropna=False))

# Ver resumen estad铆stico si aplica (solo para columnas num茅ricas planas)
print("\n Resumen estad铆stico (si aplica):")
display(df.describe())
````

### Extraer variables relevantes del diccionario (flatten columnas anidadas)

````python
# --- Paso 1: Expandir columnas anidadas ---

# Expandir columnas tipo diccionario en columnas nuevas
df_customer = pd.json_normalize(df['customer'])
df_phone    = pd.json_normalize(df['phone'])
df_internet = pd.json_normalize(df['internet'])
df_account  = pd.json_normalize(df['account'])

# Unir todo en un solo DataFrame con 'customerID' y 'Churn'
df_flat = pd.concat([df[['customerID', 'Churn']], df_customer, df_phone, df_internet, df_account], axis=1)

# Mostrar dimensiones y columnas resultantes
print(" Dimensiones del nuevo DataFrame:", df_flat.shape)
print("\n Columnas disponibles tras la expansi贸n:")
print(df_flat.columns.tolist())
````

### Variables clave para an谩lisis de churn (seg煤n el diccionario)

````python
variables_relevantes = [
    'gender',
    'SeniorCitizen',
    'Partner',
    'Dependents',
    'tenure',
    'PhoneService',
    'MultipleLines',
    'InternetService',
    'OnlineSecurity',
    'OnlineBackup',
    'DeviceProtection',
    'TechSupport',
    'StreamingTV',
    'StreamingMovies',
    'Contract',
    'PaperlessBilling',
    'PaymentMethod',
    'Charges.Monthly',
    'Charges.Total',
    'Churn'  # Variable objetivo
]

# Crear un DataFrame solo con las variables relevantes
df_churn = df_flat[variables_relevantes]

# Mostrar una muestra del DataFrame final
df_churn.head()
````

# Limpieza y Transformacion:


````python
# --- 1. Eliminar registros con 'Churn' nulo ---
df_churn = df_churn[df_churn['Churn'].notnull()]
````

````python
# --- 2. Eliminar duplicados si existen ---
df_churn = df_churn.drop_duplicates()
````

````python
# --- 3. Corregir formatos de strings (min煤sculas, espacios) ---
columnas_string = [
    'gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines',
    'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract',
    'PaperlessBilling', 'PaymentMethod', 'Churn'
]

for col in columnas_string:
    df_churn[col] = df_churn[col].astype(str).str.strip().str.lower()
````

````python
# --- 4. Reemplazar valores inconsistentes conocidos ---
valores_a_reemplazar = {
    'MultipleLines': 'No internet service',
    'OnlineSecurity': 'No internet service',
    'OnlineBackup': 'No internet service',
    'DeviceProtection': 'No internet service',
    'TechSupport': 'No internet service',
    'StreamingTV': 'No internet service',
    'StreamingMovies': 'No internet service'
}

# Reemplazar por 'No'
for col, invalido in valores_a_reemplazar.items():
    df_churn[col] = df_churn[col].replace(invalido, 'No')
````

````python
# --- 5. Convertir valores num茅ricos (por si vienen como string) ---
df_churn['SeniorCitizen'] = pd.to_numeric(df_churn['SeniorCitizen'], errors='coerce')
df_churn['tenure'] = pd.to_numeric(df_churn['tenure'], errors='coerce')
df_churn['Charges.Monthly'] = pd.to_numeric(df_churn['Charges.Monthly'], errors='coerce')
df_churn['Charges.Total'] = pd.to_numeric(df_churn['Charges.Total'], errors='coerce')
````

````python
# --- 6. Validar que no queden valores nulos tras conversi贸n ---
print("\n Valores nulos despu茅s de limpieza:")
print(df_churn.isnull().sum())
````

````python
## A煤n hay que quitar de algunas categorias " no internet service"

# --- Reemplazo de 'no internet service' y 'no phone service' ---
columnas_internet = [
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

for col in columnas_internet:
    df_churn[col] = df_churn[col].replace('no internet service', 'no')

df_churn['MultipleLines'] = df_churn['MultipleLines'].replace('no phone service', 'no')

````
