# **Challenge Telecom X**

#📌 Extracción

````python
import pandas as pd

# URL RAW del archivo JSON en GitHub
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"

# Cargar los datos desde la URL
df = pd.read_json(url)

# Mostrar las primeras filas para verificar
df.head()
````

#🔧 Transformación

````python
# Mostrar información general del DataFrame (columnas, no nulos, tipos)
print("📌 Información general del DataFrame:")
df.info()

# Tipos de datos individuales
print("\n📌 Tipos de datos por columna:")
print(df.dtypes)

# Ver las primeras filas para entender la estructura
print("\n📌 Primeras filas del dataset:")
display(df.head())

# Ver valores únicos de la variable objetivo 'Churn'
print("\n📌 Distribución de la variable 'Churn':")
print(df['Churn'].value_counts(dropna=False))

# Ver resumen estadístico si aplica (solo para columnas numéricas planas)
print("\n📌 Resumen estadístico (si aplica):")
display(df.describe())
````

### Extraer variables relevantes del diccionario (flatten columnas anidadas)

````python
# --- Paso 1: Expandir columnas anidadas ---

# Expandir columnas tipo diccionario en columnas nuevas
df_customer = pd.json_normalize(df['customer'])
df_phone    = pd.json_normalize(df['phone'])
df_internet = pd.json_normalize(df['internet'])
df_account  = pd.json_normalize(df['account'])

# Unir todo en un solo DataFrame con 'customerID' y 'Churn'
df_flat = pd.concat([df[['customerID', 'Churn']], df_customer, df_phone, df_internet, df_account], axis=1)

# Mostrar dimensiones y columnas resultantes
print("📌 Dimensiones del nuevo DataFrame:", df_flat.shape)
print("\n📌 Columnas disponibles tras la expansión:")
print(df_flat.columns.tolist())
````

### Variables clave para análisis de churn (según el diccionario)

````python
variables_relevantes = [
    'gender',
    'SeniorCitizen',
    'Partner',
    'Dependents',
    'tenure',
    'PhoneService',
    'MultipleLines',
    'InternetService',
    'OnlineSecurity',
    'OnlineBackup',
    'DeviceProtection',
    'TechSupport',
    'StreamingTV',
    'StreamingMovies',
    'Contract',
    'PaperlessBilling',
    'PaymentMethod',
    'Charges.Monthly',
    'Charges.Total',
    'Churn'  # Variable objetivo
]

# Crear un DataFrame solo con las variables relevantes
df_churn = df_flat[variables_relevantes]

# Mostrar una muestra del DataFrame final
df_churn.head()
````

#📊 Análisis de Calidad

````python
# Verificación de Calidad de Datos

# ---1. Verificar valores nulos ---
print("📌 Porcentaje de valores nulos por columna:")
print((df_churn.isnull().mean() * 100).sort_values(ascending=False))
````

````python
# --- 2. Buscar registros duplicados ---
duplicados = df_churn.duplicated().sum()
print(f"\n📌 Registros duplicados encontrados: {duplicados}")
````

````python

# --- 3. Revisar tipos de datos (por si hay errores de formato) ---
print("\n📌 Tipos de datos actuales:")
print(df_churn.dtypes)
````

````python
# --- 4. Revisar categorías únicas para columnas categóricas ---
columnas_categoricas = [
    'gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines',
    'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract',
    'PaperlessBilling', 'PaymentMethod', 'Churn'
]

print("\n📌 Valores únicos por variable categórica:")
for columna in columnas_categoricas:
    print(f"\n{columna}:")
    print(pd.unique(df_churn[columna]))
````

# Limpieza y Transformacion:


````python
# --- 1. Eliminar registros con 'Churn' nulo ---
df_churn = df_churn[df_churn['Churn'].notnull()]
````

````python
# --- 2. Eliminar duplicados si existen ---
df_churn = df_churn.drop_duplicates()
````

````python
# --- 3. Corregir formatos de strings (minúsculas, espacios) ---
columnas_string = [
    'gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines',
    'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract',
    'PaperlessBilling', 'PaymentMethod', 'Churn'
]

for col in columnas_string:
    df_churn[col] = df_churn[col].astype(str).str.strip().str.lower()
````

````python
# --- 4. Reemplazar valores inconsistentes conocidos ---
valores_a_reemplazar = {
    'MultipleLines': 'No internet service',
    'OnlineSecurity': 'No internet service',
    'OnlineBackup': 'No internet service',
    'DeviceProtection': 'No internet service',
    'TechSupport': 'No internet service',
    'StreamingTV': 'No internet service',
    'StreamingMovies': 'No internet service'
}

# Reemplazar por 'No'
for col, invalido in valores_a_reemplazar.items():
    df_churn[col] = df_churn[col].replace(invalido, 'No')
````

````python
# --- 5. Convertir valores numéricos (por si vienen como string) ---
df_churn['SeniorCitizen'] = pd.to_numeric(df_churn['SeniorCitizen'], errors='coerce')
df_churn['tenure'] = pd.to_numeric(df_churn['tenure'], errors='coerce')
df_churn['Charges.Monthly'] = pd.to_numeric(df_churn['Charges.Monthly'], errors='coerce')
df_churn['Charges.Total'] = pd.to_numeric(df_churn['Charges.Total'], errors='coerce')
````

````python
# --- 6. Validar que no queden valores nulos tras conversión ---
print("\n📌 Valores nulos después de limpieza:")
print(df_churn.isnull().sum())
````

````python
# Lista de columnas categóricas que deseamos normalizar
columnas_categoricas = [
    'gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines',
    'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract',
    'PaperlessBilling', 'PaymentMethod', 'Churn'
]

# Aplicar str.lower() y strip() para limpiar espacios y uniformar mayúsculas
for col in columnas_categoricas:
    df_churn[col] = df_churn[col].astype(str).str.strip().str.lower()
````

````python
# Verificar cómo quedó la columna 'Churn'
print("📌 Valores únicos en 'Churn' después de normalizar:")
print(df_churn['Churn'].unique())
````

````python
# --- Revisión final de limpieza---

# --- 1. Verificar valores nulos restantes ---
print("📌 Valores nulos restantes:")
print(df_churn.isnull().sum())

````

````python
# --- 2. Verificar duplicados (por si aún queda alguno) ---
duplicados_finales = df_churn.duplicated().sum()
print(f"\n📌 Registros duplicados: {duplicados_finales}")
````

````python
# --- 3. Tipos de datos tras limpieza ---
print("\n📌 Tipos de datos finales:")
print(df_churn.dtypes)
````

````python
# --- 4. Ver valores únicos por columna categórica (detecta errores tipo 'yes ', 'yees', etc.) ---
print("\n📌 Revisión de categorías únicas por columna:")
for col in df_churn.select_dtypes(include='object').columns:
    print(f"\n{col}:")
    print(sorted(df_churn[col].unique()))
````

````python
## Aún hay que quitar de algunas categorias " no internet service"

# --- Reemplazo de 'no internet service' y 'no phone service' ---
columnas_internet = [
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

for col in columnas_internet:
    df_churn[col] = df_churn[col].replace('no internet service', 'no')

df_churn['MultipleLines'] = df_churn['MultipleLines'].replace('no phone service', 'no')

````

````python
# --- Eliminar filas donde Churn esté vacío ---
df_churn = df_churn[df_churn['Churn'].str.strip() != '']
````

````python
# --- Verificación final de valores únicos tras la limpieza ---
print("\n📌 Categorías finales después de limpieza:")
for col in df_churn.select_dtypes(include='object').columns:
    print(f"\n{col}: {sorted(df_churn[col].unique())}")
````

````python
# Crear nueva columna "Cuentas_Diarias" basada en facturación mensual
df_churn['Cuentas_Diarias'] = df_churn['Charges.Monthly'] / 30

# Mostrar resumen estadístico de esta nueva variable
print("📊 Estadísticas de 'Cuentas_Diarias':")
print(df_churn['Cuentas_Diarias'].describe())

# Ver las primeras filas para validar
df_churn[['Charges.Monthly', 'Cuentas_Diarias']].head()

````

````python
# 1. Codificación binaria (yes = 1, no = 0)

# Lista de columnas binarias
binarias = [
    'Partner', 'Dependents', 'PhoneService', 'MultipleLines',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
    'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'Churn'
]

# Reemplazar 'yes' y 'no' por 1 y 0
df_churn[binarias] = df_churn[binarias].replace({'yes': 1, 'no': 0})

````

````python
#2. Renombrar columnas para mayor claridad

# Renombrar columnas si deseas tener nombres más claros
df_churn.rename(columns={
    'gender': 'Genero',
    'SeniorCitizen': 'EsAdultoMayor',
    'Partner': 'TienePareja',
    'Dependents': 'TieneDependientes',
    'PhoneService': 'ServicioTelefonico',
    'MultipleLines': 'LineasMultiples',
    'InternetService': 'TipoInternet',
    'OnlineSecurity': 'SeguridadOnline',
    'OnlineBackup': 'RespaldoOnline',
    'DeviceProtection': 'ProteccionDispositivo',
    'TechSupport': 'SoporteTecnico',
    'StreamingTV': 'TVStreaming',
    'StreamingMovies': 'PeliculasStreaming',
    'Contract': 'TipoContrato',
    'PaperlessBilling': 'FacturaDigital',
    'PaymentMethod': 'MetodoPago',
    'Charges.Monthly': 'CargoMensual',
    'Charges.Total': 'CargoTotal',
    'Churn': 'Churn',
    'Cuentas_Diarias': 'CargosDiarios'
}, inplace=True)

````

###  Conclusiones e Insights:



🔹 **Comportamiento** General de Evasión**

✅ La mayoría de los clientes no cancela el servicio, pero la proporción de cancelaciones es suficientemente significativa como para tener un impacto financiero importante.



*   📉 El churn está desbalanceado: aproximadamente 1 de cada 4 clientes cancela.
*   🔍 Clientes con menos de 12 meses tienen una probabilidad mucho mayor de cancelar.


*  🧠 Insight: Los primeros 12 meses son críticos para la fidelización.

-

🔹 Variables Numéricas

🧾 Clientes con cargos mensuales más altos muestran mayor propensión al churn.

🧠 Insight: El precio puede estar siendo percibido como alto para el valor recibido.

🪫 Los clientes que han gastado poco en total son más propensos a cancelar.

🧠 La mayoría de los que cancelan lo hacen temprano, antes de alcanzar un alto gasto acumulado.

📊Los  clientes que pagan más de $2.50 por día presentan una tasa de cancelación elevada.



🔹 Variables Categóricas

📉 Clientes con contrato month-to-month son quienes más cancelan.

🧠 Insight: La falta de compromiso de largo plazo aumenta la evasión.

🎬 Usuarios de StreamingTV y StreamingMovies tienen una ligera mayor tasa de churn.

🧠 Insight: Estos clientes pueden ser más exigentes o sensibles al valor/performance.


💻 Clientes que pagan con electronic check presentan más evasión que quienes usan pagos automáticos.

🧠 Insight: Automatizar pagos puede estar relacionado con clientes más leales o menos propensos a cancelar.


🌐 Usuarios con fiber optic muestran mayor tasa de churn que quienes tienen DSL.

🧠 Insight: Podría relacionarse con problemas técnicos, soporte o precio.




## Conclusión Estratégica
Existen dos perfiles de riesgo:

🔸 Clientes nuevos (bajo tenure, bajo cargo total)

🔸 Clientes con altos cargos (mensuales y diarios), contratos flexibles, y sin automatización de pagos



## Recomendaciones:
1. Foco en retener clientes dentro del primer año.
2. Ofrecer beneficios o descuentos a clientes con altos cargos.
3. Promover contratos anuales con beneficios adicionales.
4. Implementar modelos predictivos para anticipar la evasión.

-  Cuanto mayor el tiempo que un cliente lleva en la empresa, menor la probabilidad de que cancele.

-  Clientes nuevos tienen mayor riesgo de churn.

🔸 2. CargoTotal vs Churn: Negativa (~ -0.20)
🧠 Interpretación:

-  Clientes que han gastado mucho en total tienen menos probabilidad de abandonar.

-  Refuerza la idea de que el churn se concentra en clientes nuevos o de bajo gasto acumulado.

🔸 3. CargoMensual vs Churn: Positiva (~ +0.20)
🧠 Interpretación:

 - Clientes que pagan más cada mes tienen mayor tendencia a cancelar.

 - Puede sugerir insatisfacción percibida con el costo del servicio.


🔸 4. CargosDiarios vs Churn: Positiva (similar a CargoMensual)
🧠 Interpretación:

  - Muestra que los cargos diarios altos están relacionados con una mayor probabilidad de churn.

  - Este insight es muy útil porque refleja mejor el comportamiento de gasto cotidiano.

🔸 5. ServiciosContratados vs Churn: Leve negativa (~ -0.10 a -0.15)
🧠 Interpretación:

-  Clientes con más servicios activos tienden a permanecer más.

-  Tener más servicios podría generar mayor dependencia o satisfacción.



### 🎯 Conclusión general de la matriz de correlación:

- 🔻 **Tenure** (Antigüedad del cliente):

  Los clientes con mayor tiempo en la empresa tienen una menor probabilidad de cancelar el servicio.
  Los nuevos clientes (con pocos meses) son quienes más tienden al churn.


- 💸 **CargoTotalc(Gasto acumulado)**:

  Clientes que han gastado más dinero en total son más leales.
  El churn ocurre principalmente en clientes que aún no han generado un gasto significativo.


- 📈 **CargoMensual**:

  Existe una relación positiva entre el monto del cargo mensual y la evasión.
  Los clientes con cargos altos cancelan más, lo que puede estar relacionado con una percepción de bajo valor por el precio pagado.


- 📅 **CargosDiarios**:

  El comportamiento es similar al cargo mensual, pero visto a nivel de gasto diario.
  Un cargo diario elevado puede aumentar la sensación de “alto costo” y elevar el riesgo de churn.


- 🔌 **ServiciosContratados**:

  Hay una correlación ligeramente negativa con churn.
  Los clientes que tienen más servicios contratados (TV, soporte, respaldo, etc.) son menos propensos a irse.
  Esto podría deberse a una mayor dependencia o a una percepción de mayor valor por el paquete total.
